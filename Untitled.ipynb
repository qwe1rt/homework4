```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt


# Download the auto-mpg dataset from the UCI Machine Learning Repository
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
                'acceleration', 'model_year', 'origin', 'car_name']
# Use a raw string to avoid escape character warnings
df = pd.read_csv(url, names=column_names, sep=r'\s+', na_values='?')

# Use only continuous fields as features
continuous_features = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
X = df[continuous_features]

# Impute missing values with the mean
X = X.fillna(X.mean())

# Data standardization (optional, usually performed before clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform hierarchical clustering, removing the affinity parameter
clustering = AgglomerativeClustering(n_clusters=3, linkage='average')
labels = clustering.fit_predict(X_scaled)

# Get the mean and variance values for each cluster
cluster_means = []
cluster_variances = []
for cluster in range(3):
    cluster_data = X[labels == cluster]
    cluster_means.append(cluster_data.mean())
    cluster_variances.append(cluster_data.var())

# Use 'origin' as the category label and get the mean and variance values for each category
origin_means = df.groupby('origin')[continuous_features].mean()
origin_variances = df.groupby('origin')[continuous_features].var()

# Compare the relationship between cluster assignments and category labels
# Here, we simply determine this by looking at the proportion of different 'origin' values in each cluster
cluster_origin_counts = []
for cluster in range(3):
    cluster_data = df[labels == cluster]
    origin_counts = cluster_data['origin'].value_counts(normalize=True)
    cluster_origin_counts.append(origin_counts)

# Visualize the mean values of each cluster
cluster_means_df = pd.DataFrame(cluster_means, columns=continuous_features)
cluster_means_df.index = [f"Cluster {i}" for i in range(3)]
cluster_means_df.plot(kind='bar', figsize=(10, 6))
plt.title("Mean Values of Each Cluster")
plt.xlabel("Cluster")
plt.ylabel("Feature Value")
plt.show()

# Visualize the variance of each cluster
cluster_variances_df = pd.DataFrame(cluster_variances, columns=continuous_features)
cluster_variances_df.index = [f"Cluster {i}" for i in range(3)]
cluster_variances_df.plot(kind='bar', figsize=(10, 6))
plt.title("Variance of Each Cluster")
plt.xlabel("Cluster")
plt.ylabel("Variance Value")
plt.show()

# Visualize the mean values of each 'origin' category
origin_means.plot(kind='bar', figsize=(10, 6))
plt.title("Mean Values of Each Origin Category")
plt.xlabel("Origin")
plt.ylabel("Feature Value")
plt.show()

# Visualize the variance of each 'origin' category
origin_variances.plot(kind='bar', figsize=(10, 6))
plt.title("Variance of Each Origin Category")
plt.xlabel("Origin")
plt.ylabel("Variance Value")
plt.show()

# Visualize the proportion of different 'origin' values in each cluster
for i, counts in enumerate(cluster_origin_counts):
    counts.plot(kind='bar', figsize=(10, 6))
    plt.title(f"Proportion of Different Origins in Cluster {i}")
    plt.xlabel("Origin")
    plt.ylabel("Proportion")
    plt.show()    
```

```python
#Mean and Variance of Clusters
#Observing the "Mean Values of Each Cluster" chart: 
#The mean values of different clusters for features such as "weight" show significant differences, 
#indicating that there are distinctions among clusters in these feature dimensions. For example, 
#the mean value of Cluster 1 in the "weight" feature is much higher than that of other clusters.
#Examining the "Variance of Each Cluster" chart: 
#The variances of different clusters for various features also vary. For instance,
#Cluster 0 has a relatively large variance in the "weight" feature, 
#which means that the samples within this cluster are more dispersed in terms of the "weight" feature.

# Mean and Variance of Origin Categories
#From the "Mean Values of Each Origin Category" chart：
#Different origin categories exhibit differences in feature mean values. For example, 
#different origins show different mean values for the "weight" feature, 
#reflecting the differences in this feature among vehicles from different origins.
#Based on the "Variance of Each Origin Category" chart: 
#Different origin categories also have differences in feature variances, 
#indicating that vehicles from different origins have different degrees of dispersion in various features.

# Proportion of Origins in Clusters
#Analyzing charts such as "Proportion of Different Origins in Cluster 0",
#in Cluster 0, different origins are distributed in a certain proportion, and no single origin dominates absolutely;
#while in Cluster 1 and Cluster 2, one origin respectively dominates.

# Conclusion
#There is a certain relationship between cluster assignments and category labels (origin), but it is not completely consistent. 
#In some clusters (such as Cluster 1 and Cluster 2),
#a certain origin dominates, indicating that the clustering algorithm has a certain correlation with the origin category in these clusters;
#however, in Cluster 0, the distribution of various origins is relatively uniform, 
#suggesting that the formation of this cluster is not mainly determined by the origin. Overall, 
#the clustering results partially reflect the information of origin categories,
#but are also comprehensively affected by other features, and clusters cannot be simply divided based on the origin. 
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Obtain the Boston Housing dataset from the original link
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv(url, sep='\s+', names=column_names)

# View the basic information of the dataset
print("Basic information of the dataset:")
df.info()

# Check for missing values
print("\nStatistics of missing values:")
print(df.isnull().sum())

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Initialize variables to store the optimal k value and the corresponding silhouette coefficient
best_k = None
best_score = -1

# Iterate over the range of the number of clusters (from 2 to 6)
for k in range(2, 7):
    # Create a KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42)
    # Perform clustering
    labels = kmeans.fit_predict(scaled_data)
    # Calculate the silhouette coefficient
    score = silhouette_score(scaled_data, labels)
    print(f"k = {k}, Silhouette coefficient: {score:.4f}")
    # Update the optimal k value and the corresponding silhouette coefficient
    if score > best_score:
        best_score = score
        best_k = k

print(f"\nThe optimal k value is: {best_k}, and the corresponding silhouette coefficient is: {best_score:.4f}")

# Perform clustering with the optimal k value
best_kmeans = KMeans(n_clusters=best_k, random_state=42)
best_labels = best_kmeans.fit_predict(scaled_data)

# Calculate the means of all features in the optimal clustering
cluster_means = []
for cluster in range(best_k):
    cluster_data = df[best_labels == cluster]
    cluster_mean = cluster_data.mean()
    cluster_means.append(cluster_mean)

# Obtain the centroid coordinates
centroids = scaler.inverse_transform(best_kmeans.cluster_centers_)

# Compare the feature means and centroid coordinates
for i in range(best_k):
    print(f"\nFeature means of cluster {i}:")
    print(cluster_means[i].round(2))
    print(f"Centroid coordinates of cluster {i}:")
    print(pd.Series(centroids[i], index=df.columns).round(2))
    print("-" * 50)

# Visualize the change of the silhouette coefficient with the k value
plt.figure(figsize=(8, 6))
k_values = np.arange(2, 7)
silhouette_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_data)
    score = silhouette_score(scaled_data, labels)
    silhouette_scores.append(score)

plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette coefficient')
plt.title('Change of the silhouette coefficient with the k value')
plt.show()

# Visualize the feature means and centroid coordinates of each cluster
for i in range(best_k):
    plt.figure(figsize=(10, 6))
    feature_means = cluster_means[i]
    centroids_series = pd.Series(centroids[i], index=df.columns)

    feature_means.plot(kind='bar', label='Feature means', alpha=0.7)
    centroids_series.plot(kind='bar', label='Centroid coordinates', alpha=0.7)

    plt.title(f'Feature means and centroid coordinates of cluster {i}')
    plt.xlabel('Features')
    plt.ylabel('Values')
    plt.legend()
    plt.show()
```

```python
#Dataset Information
#The Boston Housing dataset contains 506 records and 14 fields. Most of the data types are `float64` and `int64`, 
#and there are no missing values, which is conducive to subsequent analysis.

#Optimal k - value and Clustering Effect
#After calculation, when the number of clusters ranges from 2 to 6, the optimal `k` - value is 2, 
#and the silhouette coefficient is 0.3501. A silhouette coefficient close to 1 indicates good clustering. 
#This value shows that the clustering effect is average, but within this range, `k = 2` is relatively reasonable, 
#which can also be intuitively reflected from the corresponding line chart.

#Feature Means and Centroid Coordinates
#Numerically, the two are very close, such as "CRIM" in Cluster 0 and "ZN" in Cluster 1.
#Since the centroid is calculated from the feature values of samples within the cluster, 
#and the mean is also the average of the feature values of samples within the cluster, 
#the similar principles lead to the close numerical values. In terms of meaning, 
#the feature mean reflects the average value and central tendency of samples within the cluster in the feature dimension;
#the centroid is the center of the cluster in K - Means. Through iteration,
#it minimizes the sum of the squared distances from samples to it, representing the center of the cluster from a spatial position. 
#The two describe the characteristics of the cluster from different perspectives. 
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import homogeneity_score, completeness_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the wine dataset
wine = load_wine()
X = wine.data  # Feature data
y = wine.target  # True labels

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a DataFrame with feature names
df = pd.DataFrame(X_scaled, columns=wine.feature_names)
df['true_label'] = y

print(f"Dataset shape: {df.shape}")
print(f"Distribution of true classes:\n{df['true_label'].value_counts()}")
print(df.head())

# Set the number of clusters to 3
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Calculate the cluster centers
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
centroid_df = pd.DataFrame(centroids, columns=wine.feature_names)

print(f"Cluster centers:\n{centroid_df.round(2)}")
print("\nCross-table of true classes and clustering results:")
print(pd.crosstab(df['true_label'], df['cluster'], rownames=['True Classes'], colnames=['Clusters']))

# Calculate homogeneity and completeness
homogeneity = homogeneity_score(y, df['cluster'])
completeness = completeness_score(y, df['cluster'])

print(f"Homogeneity score: {homogeneity:.4f}")
print(f"Completeness score: {completeness:.4f}")

# Visualize the homogeneity and completeness metrics
plt.figure(figsize=(10, 6))
metrics = ['Homogeneity', 'Completeness']
scores = [homogeneity, completeness]

sns.barplot(x=metrics, y=scores)
plt.ylim(0, 1)
plt.title('Clustering Quality Evaluation Metrics')
plt.ylabel('Score')

for i, v in enumerate(scores):
    plt.text(i, v + 0.01, f'{v:.4f}', ha='center')

plt.show()

# Further visualize the relationship between clustering results and true labels
# Plot the distribution of true labels in each cluster
plt.figure(figsize=(12, 8))
sns.countplot(x='cluster', hue='true_label', data=df)
plt.title('Distribution of True Labels in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Number of Samples')

plt.show()
```

```python
#Homogeneity: It assesses the singularity of the true classes of samples within a cluster.
#The closer the value is to 1, the more samples within the cluster belong to the same true class.
#In this case, the value is 0.8788, indicating that the clustering has a relatively good effect in distinguishing true classes,
#and most samples within the clusters come from a single class.

#Completeness: It measures the degree of aggregation of samples of the same true class. The closer the value is to 1, 
#the better the aggregation of samples of the same class. Here, the value is 0.8730, 
#suggesting that most samples of the true classes are well - aggregated,
#but some samples of true class 1 are scattered (as shown in the cross - tabulation). 
```
